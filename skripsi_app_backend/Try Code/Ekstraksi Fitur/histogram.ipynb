```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('../../Code/Dataset/preprocessed_roi_augmentation/5_label/dataset_testing/bercak_daun/data ke-9.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Ubah ke mode warna RGB
```

```python
# Tentukan jumlah "bin" yang diinginkan, misalnya 16 atau 32
bins = 16

# Hitung histogram warna menggunakan jumlah "bin" yang ditentukan
hist_red = cv2.calcHist([image], [0], None, [bins], [0, 256])  # Saluran merah
hist_green = cv2.calcHist([image], [1], None, [bins], [0, 256])  # Saluran hijau
hist_blue = cv2.calcHist([image], [2], None, [bins], [0, 256])  # Saluran biru
```

```python
# Tampilkan gambar dan histogram dalam subplot terpisah untuk masing-masing saluran warna
plt.figure(figsize=(15, 5))

# Subplot untuk tampilan gambar
plt.subplot(1, 4, 1)
plt.imshow(image)
plt.axis('off')  # Matikan sumbu x dan y
plt.title('Gambar')

# Subplot untuk tampilan histogram saluran merah
plt.subplot(1, 4, 2)
plt.plot(hist_red, color='red')
plt.xlim([0, bins - 1])
plt.xlabel('Nilai Warna')
plt.ylabel('Frekuensi')
plt.title('Histogram Merah')

# Subplot untuk tampilan histogram saluran hijau
plt.subplot(1, 4, 3)
plt.plot(hist_green, color='green')
plt.xlim([0, bins - 1])
plt.xlabel('Nilai Warna')
plt.ylabel('Frekuensi')
plt.title('Histogram Hijau')

# Subplot untuk tampilan histogram saluran biru
plt.subplot(1, 4, 4)
plt.plot(hist_blue, color='blue')
plt.xlim([0, bins - 1])
plt.xlabel('Nilai Warna')
plt.ylabel('Frekuensi')
plt.title('Histogram Biru')

plt.tight_layout()  # Rapihkan tata letak subplot
plt.show()
```

```python
# Tampilkan gambar dan histogram dalam dua subplot terpisah
plt.figure(figsize=(12, 6))

# Subplot untuk tampilan gambar
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.axis('off')  # Matikan sumbu x dan y

# Subplot untuk tampilan histogram
plt.subplot(1, 2, 2)
# Histogram untuk saluran merah
plt.plot(hist_red, color='red', label='Red')
# Histogram untuk saluran hijau
plt.plot(hist_green, color='green', label='Green')
# Histogram untuk saluran biru
plt.plot(hist_blue, color='blue', label='Blue')

# Sesuaikan batas sumbu x dengan jumlah "bin" yang digunakan
plt.xlim([0, bins - 1])
plt.xlabel('Nilai Warna')
plt.ylabel('Frekuensi')
plt.title('Histogram Warna ({} Bins)'.format(bins))
plt.legend()

plt.show()
```
